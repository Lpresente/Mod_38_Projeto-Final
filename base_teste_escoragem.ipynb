```python
import pandas as pd

# 1. Carregar a base de dados original (.ftr)
df = pd.read_feather('credit_scoring.ftr')

# 2. Pegar apenas as primeiras 50 linhas para servir de teste
# (Você pode mudar o número 50 para o quanto quiser)
df_teste = df.head(50).copy()

# 3. Remover a coluna 'mau' (target), pois no cenário real de escoragem 
# o Streamlit não deve receber a resposta se o cliente é bom ou mau.
if 'mau' in df_teste.columns:
    df_teste = df_teste.drop(columns=['mau'])

# 4. Salvar essas linhas em um arquivo CSV chamado 'base_teste_escoragem.csv'
# Configurado com index=False para não criar uma coluna de índices fantasma no CSV
df_teste.to_csv('base_teste_escoragem.csv', index=False)

print("Arquivo 'base_teste_escoragem.csv' gerado com sucesso!")
```
